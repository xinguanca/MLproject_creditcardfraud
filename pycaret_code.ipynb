# 1. Classification model test

```python
from sklearn.metrics import average_precision_score
from pycaret.classification import *
```

```python
clf = setup(
    df_model, target = 'Class', fold=5, train_size=0.8,
    use_gpu=True, session_id=1
)
add_metric('PRC_AUC', 'PRC_AUC', average_precision_score)
```

```python
bestModel = compare_models()
```

```python
plot_model(bestModel, plot='pr')
```

```python
get_metrics()
```
